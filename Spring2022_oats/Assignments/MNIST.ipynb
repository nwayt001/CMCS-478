# ML Demo

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
# Load digits dataset
digits = datasets.load_digits()
```

```python
# Get attribute vector
X = digits['data']

# Get class labels
y = digits['target']
```

```python
# What does an instance contain?
X[0]
```

```python
# What do they look like?
for i in range(10):
    img = (np.reshape(X[i], (8, 8))).astype(np.uint8)
    plt.imshow(img, interpolation='nearest', cmap = 'gray')
    plt.show()
```

```python
# What are the true class labels?
y[:10]
```

```python
# Initialize a model
clf = LogisticRegression(C = 0.01)
```

```python
# Fit the model to the data
clf.fit(X, y)
```

```python
# Look at some sample predictions
clf.predict(X[:10])
```

```python
# Check accuracy
right = [pred == true for pred, true in zip(clf.predict(X), y)]
print('Accuracy = %.2f' % (100 * (sum(right) / len(y))))
```
